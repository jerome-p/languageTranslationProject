```python
#Loading data usign pandas
eval_prec = pd.read_csv(folder_name+"/eval_precision.csv")
eval_bleu1 = pd.read_csv(folder_name+"/eval_bleu1.csv")
eval_bleu2 = pd.read_csv(folder_name+"/eval_bleu2.csv")
eval_f1 = pd.read_csv(folder_name+"/eval_f1.csv")
eval_recall = pd.read_csv(folder_name+"/eval_recall.csv")
eval_acc = pd.read_csv(folder_name+"/eval_accuracy.csv")
t5_train_loss = pd.read_csv(folder_name+"/train_loss.csv")
eval_loss = pd.read_csv(folder_name+"/eval_loss.csv")
# metrics = [eval_prec,eval_bleu1,eval_bleu2,eval_f1,eval_recall,eval_acc,t5_train_loss,eval_loss]
# eval_prec, eval_loss
```

```python
# Defingin the funciton
def create_line_plot(data,xlabel="", ylabel="", title='',model_name=''):
    """
    data: DataFrame
    xlabel: str
    ylabel: str
    title: str
    Returns None
    """
    plt.figure(figsize=[15,8])
    plt.plot(data, label=title)

    # Labelling
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(facecolor='xkcd:light grey')
    plt.savefig('graphs2/'+model_name+'_'+title+'_graph.png', bbox_inches='tight', dpi=300)
    plt.show()

    return 
```
