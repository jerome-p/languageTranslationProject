```python

def create_line_plot(data,xlabel="", ylabel="", title=''):
    """
    data: DataFrame
    xlabel: str
    ylabel: str
    title: str
    Returns None
    """
    plt.figure()
    plt.plot(data, label=title)

    # Labelling
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(facecolor='xkcd:light grey')
    plt.show()

    return 
```
